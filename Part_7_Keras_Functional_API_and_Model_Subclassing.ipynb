# Part 7: Keras Functional API and Model Subclassing

In this notebook, we'll explore more advanced ways to build models in TensorFlow 2.x beyond the simple `Sequential` API.

1. **Sequential API**: Best for a plain stack of layers.
2. **Functional API**: Best for complex models with multiple inputs, outputs, or shared layers.
3. **Model Subclassing**: Best for research and fully customizable models.

```python
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
```

## 1. Functional API
The Functional API treats layers as functions. You define an input tensor and then call layers on it.

```python
def build_functional_model():
    inputs = layers.Input(shape=(32,))
    x = layers.Dense(64, activation='relu')(inputs)
    x = layers.Dense(64, activation='relu')(x)
    outputs = layers.Dense(10, activation='softmax')(x)

    model = tf.keras.Model(inputs=inputs, outputs=outputs, name="functional_model")
    return model

func_model = build_functional_model()
func_model.summary()
```

## 2. Model Subclassing
By subclassing `tf.keras.Model`, you define your layers in `__init__` and the forward pass in `call`.

```python
class MyCustomModel(tf.keras.Model):
    def __init__(self):
        super(MyCustomModel, self).__init__()
        self.dense1 = layers.Dense(64, activation='relu')
        self.dense2 = layers.Dense(64, activation='relu')
        self.dense3 = layers.Dense(10, activation='softmax')

    def call(self, inputs):
        x = self.dense1(inputs)
        x = self.dense2(x)
        return self.dense3(x)

subclassed_model = MyCustomModel()
# Note: Subclassed models need to be called on data or built before summary() works
subclassed_model.build(input_shape=(None, 32))
subclassed_model.summary()
```
